# Type of join 
Its not necesary for both of the joining column to have save values. you can still join them. there are four type of join
- **Inner join:** it joins only those values which are present on in both columns. 
- **Outer join:** it joins only those values with all values but if the value doesnt have respecting coulmn vlaue it will store as *NaN*. 
- **Lest join:** it joins all column from left column but matching values from right column. 
- **Right join:** it joins all colum from right column but matching values from left column.

```python
a = pd.DataFrame({
    "color": ['red', 'yellow', 'gareen'],
    "intencity": [1,5,4]
})
b = pd.DataFrame({
    "color": ['red', 'black', 'gareen'],
    "size": ['M','S','L']
})
display(a)
display(b)
```

## Inner Join

```python
ab_inner = pd.merge(a, b, how='inner')
ab_inner
```

## Outer Join

```python
ab_outer = pd.merge(a, b, how='outer')
ab_outer
```

## Left Join

```python
ab_left = pd.merge(a, b, how='left')
ab_left
```

## Right Join

```python
ab_right = pd.merge(a, b, how='right')
ab_right
```
